## Step 3️⃣: Cleaning the Data

```python
# Function that selects columns with a number of non-null values greater than m
def df_non_null_columns(df,m):
    
    lg = []
    
    for item in list_column_names:
        
        is_nn_list = df[item].notnull()
        nn = sum(is_nn_list)
        
        if nn > m:
            lg.append(item)
    
    return df[lg]
        
```

```python
# Function that selects only numerical columns 
def df_numerical_columns(df):
    
    df_nc = df.select_dtypes(np.number)
    
    return df_nc
```

```python
# Select only colums with more than 1000 non-null values
df1 = df_non_null_columns(df,1000)
df1
```

```python
# List of columns names with more than 1000 non-null values
list_column_names_non_null = list(df1.columns)
```

```python
df1.info();
```

```python
# Drop NaN rows 
df1d = df1.dropna()
df1d = df1d.reset_index(drop=True)
df1d.info()
```

```python
# Select only columns with numerical values
df1n = df_numerical_columns(df1d)
df1n
```

```python
# List of columns names with numerical values
list_column_names_numerical = list(df1n.columns)
```

```python
# List of columns names with categorical values
list_column_names_cat = [item for item in list_column_names_non_null if item not in list_column_names_numerical]
list_column_names_cat;
```

```python
# Function that adds dummies of selected categorical feature to dataframe
def add_1cat_dummies(df,name):
    
    dummies = pd.get_dummies(df[name])
    dfnn = df_numerical_columns(df)
    dfat = dfnn.join(dummies)
    
    return dfat
```

```python
# Import random generator
from random import randrange
```

```python
# Select a random categorical colum and add dummies to the numerical columns
rname = list_column_names_cat[randrange(len(list_column_names_cat))]
df1nc = add_1cat_dummies(df1d,rname)
```

# 🟥🟥Phase B: What Will Happen? (AI Models)
The objective in this phase is using Machinea Learning algorithms (through python libraries) to build a predictive model that can get a few inout features such as the number of bedrooms, the number of rooms, income level within the area and tell us if this given house is likly to be priced equal or greater than $1M dollars!

## Step 5️⃣: Predictive Modeling

```python
# Remove Id column
del df1nc['Id']
```

```python
# let's take a look at all available columns
df1nc.columns
```

```python
# Define Features and Target based on target name
def features_and_target(df,target_name):
    
    cols = list(df.columns.values) # Make a list of all of the columns in the df
    cols.pop(cols.index(target_name)) # Remove target name from list
    X = df[cols]
    y = df[target_name]
    
    return X, y
    
```

```python
X, y = features_and_target(df1nc,'SalePrice')
```

```python
# Import the required library for train-test split
from sklearn.model_selection import train_test_split

# Splitting entire data randomly into test (30%) and trainig (70%)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
# Import the required library for ML model
from sklearn.ensemble import RandomForestRegressor

# Use RandomForest model
mrfr = RandomForestRegressor() 

# Train the model
mrfr.fit(X_train, y_train) #fit the model to the data
```

```python
# Display accuracy of the model on training and testing data
print('Accuracy of RF regression on training set: {:.2f}'.format(mrfr.score(X_train, y_train))) #accuracy on training data
print('Accuracy of RF regression on test set: {:.2f}'.format(mrfr.score(X_test, y_test))) #accuracy on test data
```

# 🟥🟥🟥 Phase C: Are You Sure? (Model Evaluation)
The objective of this phase is evaluating the model and make sure it works! There are various ways and metrics to evaluate and quantify the performance of the `Regression` model. 

## Step 6️⃣: Evaluation metrics
In `Regression` type of problems our goal is to make a prediction about a target-label which takes continue numerical values. For example, predicting the price of a house which can be any number (sometimes within a specific range).

Typical metrics used to evaluate this kind of models are: 1) Mean Absolute Error, 2) Mean Squared Error, 3) Root Mean Squared Error and 4) R2 score.

MAE (Mean absolute error) represents the difference between the original and predicted values extracted by averaged the absolute difference over the data set.

MSE (Mean Squared Error) represents the difference between the original and predicted values extracted by squared the average difference over the data set.

RMSE (Root Mean Squared Error) is the error rate by the square root of MSE.

R2 (Coefficient of determination) represents the coefficient of how well the values fit compared to the original values. The value from 0 to 1 interpreted as percentages. The higher the value is, the better the model is.

```python
# Import metrics
from sklearn import metrics
from sklearn.metrics import mean_absolute_error as MAE
from sklearn.metrics import mean_squared_error as MSE
from sklearn.metrics import r2_score as R2
```

```python
# Predict and show evaluation metrics for our model
y_pred = pd.Series(mrfr.predict(X_test))
print('RFR Mean Absolute Error:', int(MAE(y_test, y_pred)))
print('RFR Mean Squared Error:', int(MSE(y_test, y_pred)))
print('RFR Root Mean Squared Error:', int(np.sqrt(MSE(y_test, y_pred))))
print('RFR R2 score:', R2(y_test, y_pred))
```

```python
print( int(np.sqrt(MSE(y_test, y_pred)))/R2(y_test, y_pred))
```
